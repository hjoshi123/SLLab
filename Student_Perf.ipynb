```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('StudentsPerformance.csv')
```

```python
df.head(20)
```

```python
df1 = df.drop(['lunch', 'test preparation course'], axis=1)
```

```python
df1["parental level of education"] = df["parental level of education"].fillna("high school")
```

```python
df1.head(20)
```

```python
df1["race/ethnicity"] = df1['race/ethnicity'].map({
    "group A": "Asian Students",
    "group B": "African Students",
    "group C": "Afro-Asian Students",
    "group D": "American Students",
    "group E": "European Students"
})
```

```python
df1.head(20)
```

```python
ax = sns.countplot(x="test preparation course", hue="gender", palette="Set1", data=df)
ax.set(title="Test Preparation", xlabel="Course", ylabel="Total")
plt.show()
```

```python
ax = sns.countplot(x="race/ethnicity", hue="gender", palette="Set1", data=df1)
ax.set(title="Students according to each group", xlabel="Ethnicity", ylabel="Total")
plt.show()
```

```python
marks_intervals = [0, 40, 50, 60, 75]
categories = ['failed', 'second class', 'first class', 'distinction']

df1['MarksCategoriesMath'] = pd.cut(df1['math score'], marks_intervals, labels=categories)
ax = sns.countplot(x="MarksCategoriesMath", hue="gender", palette="Set1", data=df1)
ax.set(title="Math Marks", xlabel="Marks Intervals", ylabel="Total")
plt.show()
```

```python
df1['MarksCategoriesReading'] = pd.cut(df['reading score'], marks_intervals, labels=categories)
ax = sns.countplot(x='MarksCategoriesReading', hue='gender', palette='Set1', data=df1)
ax.set(title='Reading Marks', xlabel='Marks Intervals', ylabel='Total')
plt.show()
```

```python
df1['MarksCategoriesWriting'] = pd.cut(df1['writing score'], marks_intervals, labels=categories)
ax = sns.countplot(x="MarksCategoriesWriting", hue="gender", palette="Set1", data=df1)
ax.set(title="Writing Marks", xlabel="Marks Intervals", ylabel="Total")
plt.show()
```
